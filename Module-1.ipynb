<h3>Sample Cipher Crack Procedure: </h3>

```python
def decypher_value(encrypted_value, initial_shift, increment, period_length):
    inner_disc = list(range(0,256))
    inner_disc = list(np.roll(inner_disc, initial_shift))
    current_inc = 0
    decoded_value = list()
    for character in encrypted_value:
        decoded_value.append(inner_disc.index(character))
        if current_inc == period_length:
            inner_disc = list(np.roll(inner_disc, increment))
            current_inc = 0
        current_inc = current_inc+1
    return decoded_value
```

```python
video = encrypted_video
text = encrypted_text
def crack_alberti(encrypted_text):
    # to ease students into the modules, this one has them do a simple brute force attack. the other ones won't be so easy ;)
    # will throw an error if value is not valid text, so we'll use a try catch to handle this
    possible_combinations = list(itertools.product(list(range(0,26)), repeat=3))
    for combination in possible_combinations:
        initial_shift = combination[0]
        increment = combination[1]
        period_length = combination[2]
        decyphered_value = decypher_value(encrypted_text, initial_shift, increment, period_length)
        try:
            print(str(bytes(decyphered_value), 'utf-8'), '(', initial_shift, ',',increment, ',', period_length,')')
        except:
            continue
```
